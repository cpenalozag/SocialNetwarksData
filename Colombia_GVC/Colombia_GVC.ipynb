# Análisis GVC Colombia

```
import matplotlib.pyplot as plt
from operator import itemgetter
import networkx as nx
import pandas as pd
import difflib
import operator
import math
%matplotlib inline
```

## Importar redes

```
networks = [('Chemicals and chemical products', nx.read_graphml('Chemicals and chemical products.graphml')),
            ('Education services', nx.read_graphml('Education services.graphml')),
            ('Non-metallic mineral products', nx.read_graphml('Non-metallic mineral products.graphml'))]
```

## Análisis de redes

```
for i, G in networks:
  countries_va = []
  count = 0
  for n in G.nodes():
    edges = [e['weight'] for u,v,e in G.edges(data=True) if u == n]
    va=sum(edges)
    if len(edges)>1:
      count = count + 1
      countries_va.append((n,va))
  if (count>1):
    print(i,':')
    print('Densidad: ', nx.density(G))
    print('Clustering promedio', nx.average_clustering(G))
    print('Max:', max(countries_va,key=itemgetter(1)))
    print('Min:', min(countries_va,key=itemgetter(1)))
    avg = sum([item[1] for item in countries_va])/len([item[1] for item in countries_va])
    print('Average:', avg)
    col = [item for item in countries_va if item[0] == 'COLOMBIA'][0]
    print('Colombia:', col[1])
    print('Encima de promedio:', avg<col[1])
    print('n:', count, '\n')
    
    # Eigenvector centrality
    print('Ranking eigenvector centrality:')
    ev_centrality=nx.eigenvector_centrality(G,max_iter=1000, weight='weight') # compute eigenvector centrality

    for index, country in enumerate(sorted(ev_centrality.items(), key=lambda x: x[1], reverse=True)[:5]):
      print(index+1,country[0],country[1])
  

    print('\n')
  
    # betweenness centrality
    print('Ranking betweenness centrality:')
    betweenness_centrality = nx.betweenness_centrality(G, weight='weight')

    for index, country in enumerate(sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]):
      print(index+1,country[0],country[1])
    

    
    print('\n---------------------------------')

  
```
